### 1.3 Reproducible Pipelines with DVC & `params.yaml`

Data Version Control (DVC) enables reproducible data pipelines. Instead of running notebooks manually in order, we write pipeline stages in `dvc.yaml` and keep all configuration parameters in `params.yaml`.

- `dvc.yaml`: Tracks stage dependencies (`deps`), command to run (`cmd`), and output artifacts (`outs`).
- `params.yaml`: Centralized configuration (epochs, learning rate, data paths) used by stages.

Let's write a small YAML config file and load it using Python to see how configs are managed at runtime.

````python
import yaml

config_content = '''
model:
  learning_rate: 0.001
  epochs: 10
  batch_size: 32
'''

config = yaml.safe_load(config_content)
print('Loaded Parameters:')
print(f'- Learning Rate: {config["model"]["learning_rate"]}')
print(f'- Epochs: {config["model"]["epochs"]}')
````

## Section 2: Coding Agents & AI Ecosystem

### 2.1 Introduction to Agentic Coding Assistants

Unlike basic chat models (where you copy-paste code snippets manually), **Coding Agents** live directly inside your workspace, terminal, or IDE. They interact with your file system, execute terminal commands, run testing suites, and solve tasks autonomously.

Key tools include:
- **Claude Code**: Anthropic's CLI-based coding agent designed to run terminal commands, write tests, and edit project code directly.
- **Antigravity**: Google DeepMind's agentic coding system that handles codebase exploration, file edits, and execution tasks.
- **OpenAI Codex**: The foundation behind GitHub Copilot, facilitating real-time code autocomplete and inline suggestions.

Because they can search the directory, write files, and run terminal commands, these tools are vastly superior at executing changes than humans manually copying and pasting code blocks.

### 2.2 Connection: Project Workspace & Agent Context

An agent operates at two layers of understanding: the **Domain/Subject Level** and the **Operational/Execution Level**.

```
+------------------------------------------------------------------+
|                     Project Workspace Docs                       |
|  - Domain Context (Macroeconomics, Public Policy, Class schedule)|
|  - System Software Requirements, Notebooks & docs/ references    |
|  - Individual & Group Assignments (Deliverables)                 |
+-------------------------------+----------------------------------+
                                | (Translates Domain into Action)
                                v
+------------------------------------------------------------------+
|                     .agents/ Tooling                             |
|  - Strict Execution Guardrails & Rules                           |
|  - Git Commit Conventions (ALL-CAPS Verbs, Atomic commits)       |
|  - Step-by-Step Approval Gates (Human-in-the-Loop)               |
+------------------------------------------------------------------+
```

#### Why this Connection is Crucial
1. **Context Density**: By matching these two sources, an agent understands the boundaries of its current week's assignment or project scope, preventing it from making unrelated file changes.
2. **AI-Ready Hierarchy**: Standardizing project maps and data layouts (e.g., Cookie Cutter data layout) allows parsing tools to quickly ingest codebase layouts without wasteful tree crawls.
3. **Human-in-the-Loop (HITL) Gatekeeping**: Coding agents must never bypass human verification. The workflows mandate explicit validation gates, such as presenting proposed commit messages for approval first.

### 2.3 API Subscription Ecosystem & Platforms

Developing with LLMs requires choosing the right model providers, APIs, and subscription tiers. Important players include:
