### 1.3.1 Function to download data from the world bank WDI

### Just trust me for now!
Don't worry if the code block below looks like an alien language. We are defining a custom function to grab web data. Just execute the cell to see the data, and we will break down exactly how variables and data types work in the next notebook.

```python
import requests
import pandas as pd
import io

def download_worldbank(indicator, countries, date_start, date_end):
    url_base = 'https://api.worldbank.org/v2/'
    country_codes = ';'.join(countries)
    url = f'{url_base}country/{country_codes}/indicator/{indicator}?date={date_start}:{date_end}&per_page=30000'
    response = requests.get(url)
    # Use io.BytesIO to pass the bytes content as a file-like object to read_xml
    df = pd.read_xml(io.BytesIO(response.content))
    return df
```

```python
# Example 1
data = download_worldbank(
    indicator = 'NY.GDP.PCAP.CD' , 
    countries = ['US', 'CA', 'MX', 'JP'],  
    date_start = '2021', 
    date_end = '2023'
)
data.head(2)
```
